```python
# Imports and display settings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Defaults
pd.options.display.max_columns = 100
pd.options.display.max_rows = 50
sns.set(style="whitegrid") #  seaborn default
%matplotlib inline
```

```python
# Column names for UCI dataset
column_names = [
    "age",
    "workclass",
    "fnlwgt",
    "education",
    "education_num",
    "marital_status",
    "occupation",
    "relationship",
    "race",
    "sex",
    "capital_gain",
    "capital_loss",
    "hours_per_week",
    "native_country",
    "income"
]

df = pd.read_csv(r"C:/Users/darre/adult_census_eda/data/adult.data",
                 header=None,
                 names=column_names,
                 na_values="?",
                 skipinitialspace=True)
```

```python
df.shape
```

```python
# Quick checks
display(df.head(10))
df.info()
df.describe(include="all").T
```

```python
# Column Type Detection
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
object_cols = df.select_dtypes(include=["object","category"]).columns.tolist()
print("numeric:", numeric_cols)
print("categorical/object:", object_cols)
```

```python
# Missing placeholders and unique counts
# Count of "?" occurrences per column
q_counts = {col: (df[col] == "?").sum() for col in df.columns if df[col].dtype == "object"}
display(pd.Series(q_counts).sort_values(ascending=False))

# Unique counts for cat. cols
df[object_cols].nunique().sort_values(ascending=False)
```

```python
# Quick data dictionary
data_dict = pd.DataFrame({
    "column": df.columns,
    "dtype": [str(df[c].dtype) for c in df.columns],
    "n_unique": [df[c].nunique(dropna=False) for c in df.columns],
    "sample_values": [str(df[c].dropna().unique()[:5]) for c in df.columns]
})
data_dict
```

```python
# Ensuring correct data loads
plt.figure(figsize=(8,4))
df['age'].hist(bins=30)
plt.title("Age distribution")
plt.xlabel("Age")
plt.ylabel("count");

plt.figure(figsize=(10,4))
df['education'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Education Levels")
plt.xlabel("Education")
plt.ylabel("Count");
```

** Day 2: Data Clenaing Notes **
- Categorical Columns: workclass, education, marital_status, occupation, relationship, race, sex, native_country, income
- Numerical Columns: age, fnlwgt, education_num, capital_gain, capital_loss, hours_per_week, income

```python
# Convert columns to correct types
for col in object_cols:
    df[col] = df[col].astype("category")

display(df.dtypes)

# Replace "?" with NaN
df.replace("?", np.nan, inplace=True)
df.isna().sum()
```

```python
# Category counts for categorical columns
for col in object_cols:
    print(f"\n{col} value counts:")
    print(df[col].value_counts())

df.describe()
```
